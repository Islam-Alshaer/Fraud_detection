```python
import json
#in this notebook we will analyze the results of the experiments
#open results_log.jsonl
results = []
with open('results_log.jsonl', 'r') as f:
    results = json.load(f)
```

```python
#best 3 in precision
best_precision = sorted(results, key=lambda x: x['evaluation']['precision'], reverse=True)[:3]
print("Best precision:", json.dumps(best_precision, indent=4))
```

```python
#best 4 in recall
best_recalls = sorted(results, key=lambda x: x['evaluation']['recall'], reverse=True)[:4]
print("Best recall: ", json.dumps(best_recalls, indent=4))
```

```python
#best 3 in f2 score
best_f2 = sorted(results, key=lambda x: x['evaluation']['F2 Score'], reverse=True)[:3]
print("Best 3 in f2 score: ", json.dumps(best_f2, indent=4))
```

```python
#best 3 in f1 score
best_f1 = sorted(results, key=lambda x: x['evaluation']['F1 Score'], reverse=True)[:3]
print("Best 3 in f1 score: ", json.dumps(best_f1, indent=4))
```

```python
#best 3 in f0.5 score
best_f_1_1 = sorted(results, key=lambda x: x['evaluation']['F1/2 Score'], reverse=True)[:3]
print("Best 3 in f1/2 score: ", json.dumps(best_f_1_1, indent=4))
```

```python
#lowest 3 in FPR
lowest_fpr = sorted(results, key=lambda x: x['evaluation']['False Positive Rate'])[:3]
print("Lowest 3 in FPR: ", json.dumps(lowest_fpr, indent=4))
```
